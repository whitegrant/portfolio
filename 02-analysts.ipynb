```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('http://diether.org/prephd/02-mstk_analysts.csv',parse_dates=['caldt'])
```

```python
# 1. Print out the first 10 observations of the data.
df.head(10)
```

```python
# 2. Create a new column in the dataframe the contains the natural log of (1 + analysts).
df['log_analysts'] = np.log(1 + df['analysts'])
df
```

```python
# 3. During June of 2020, what is the price of Hormel's stock? Note, the ticker symbol for Hormel is HRL.
june = df[df['caldt'].dt.month == 6]
hrl_june_prc = june[june['ticker'] == 'HRL']['prc']
hrl_june_prc
```

Hormel's stock price is $48.27 in June of 2020.

```python
# 4. What is the average number of analysts covering Tesla this year? Note, the ticker symbol for Tesla is TSLA.
df[df['ticker'] == 'TSLA']['analysts'].mean()
```

```python
# 5. Creat a new column in the dataframe that is True if the number of analysts is greater than 10 and False otherwise.
df['gt_10_analysts'] = df['analysts'] > 10
df
```

```python
# 6. Compute the average market-cap (the column named me) for observations with more than 10 analysts and observations with less than 10 analysts.
df.groupby('gt_10_analysts')['me'].mean()
```

The average market-cap for observations with more than 10 analysts is about $35,443,189,887.

The average market-cap for observations with less than 10 analysts is about $1,534,442,337.

```python
# 7. Compute the number of stocks in the dataframe by month.
df.groupby('caldt')['ticker'].count()
```

```python
# 8. Compute the aggregate market-cap of all stocks in the dataframe by month.
df.groupby('caldt')['me'].sum()
```

```python
# 9. Create a new dataframe (call it sub) that contains all the observations of Google (ticker=MSFT) and Microsoft (MSFT).
tckr_list = ['GOOG', 'GOOGL', 'MSFT']
sub = df[df['ticker'].isin(tckr_list)]
sub
```
